#### **Sixth step**: data cleaning.

```python
# Specify new column names

new_cols_dict = {
    'timestamp' : 'time',
    'cnt' : 'count',
    't1' : 'temp_real_C',
    't2' : 'temp_feels_like_C',
    'hum' : 'humidity_percent',
    'wind_speed' : 'wind_speed_kph',
    'weather_code' : 'weather',
    'is_holiday' : 'is_holiday',
    'is_weekend' : 'is_weekend',
    'season' : 'season'
}

# Rename the columns

bikes.rename(new_cols_dict, axis = 1, inplace = True)
```

```python
# Change the humidity values to percentage (i.e., a value between 0 and 1)

bikes.humidity_percent = bikes.humidity_percent / 100
```

```python
# Create a season dictionary to link the code to its description

season_dict ={
    '0.0' : 'Spring',
    '1.0' : 'Summer',
    '2.0' : 'Autumn',
    '3.0' : 'Winter'
}

# Create a weather dictionary to link the code to its description

weather_dict ={
    '1.0' : 'Clear',
    '2.0' : 'Scattered clouds',
    '3.0' : 'Broken clouds',
    '4.0' : 'Cloudy',
    '7.0' : 'Rain',
    '10.0' : 'Rain with thunderstorm',
    '26.0' : 'Snowfall'
}

# Change the season column data type to string
bikes.season = bikes.season.astype('str')

# Map the values 0-3 to the actual written seasons
bikes.season = bikes.season.map(season_dict)

# Change the weather column data type to string
bikes.weather = bikes.weather.astype('str')

# Map the values to the actual written weathers
bikes.weather = bikes.weather.map(weather_dict)
```
